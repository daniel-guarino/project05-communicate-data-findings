## Investigation Overview

- The goal of this work is to show why people ask for a loan at Prosper, what's the people most common reasons to ask a loan and how the income range affects the Interest Rate. It also shows how the interest is affected by a rating score and loan term.


## Dataset Overview

- This is a reduced dataset (12 variables out 81) about personal loans, where we have important attributes to describe a loan status like the Amount of money asked, the Term (in months), the Interest Rates, the Rating Scores among others attributes. This data was provided by Prosper.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_loan = pd.read_csv('prosperLoanDataReduced.csv')
```

## Loan Term in Months

Here it's possible to notice that most people ask a loan for a 36 month term, followed by 60 and 12 months. The last one with very small number of loans. The chart is not sorted by the time in months, but according to the frequency in the descending order.

```python
df_loan['Term'].value_counts().plot(kind='bar');
plt.title('Loan Term (in Months)')
plt.xlabel('Time in months')
plt.ylabel('Frequency of loans');
```

## Loan Reasons

It's important to point that the most common reason to ask a loan is to pay another debt, since the most common is Debt Consolidation as can be seen in the following chart. It's also important to say that we conclude the most common cause without knowing the reason for almost 20,000 loans. The chart below shows the six most common reasons (including the number of unavailable data).

```python
df_loan['ListingCategory'].value_counts()[0:6].plot(kind='bar');
plt.title('Reasons people use to ask a loan')
plt.xlabel('Reason')
plt.ylabel('Frequency of loans');
```

## Term, Loan Amount and Interest Rate

- Loan Amounts higher than USD 25,5k are always made by people with the highest income range USD 100k.
- Most of unemployed people ask a 60 month loan term.
- The most common income range people (USD 25k to USD 50k) ask for a 12 loan term.
- People with no income asks more for a 36 month loan term.

```python
g = sns.FacetGrid(data = df_loan, col = 'Term', margin_titles = True, hue = 'IncomeRange')
(g.map(plt.scatter, 'LoanOriginalAmount', 'BorrowerRate').add_legend());
```
